```python
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sb
url='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
df=pd.read_csv(url,sep=',')
df.head()
```

```python
# 1.    Delete the Unnamed 0 column
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()#checks if the column has been deleted
```

```python
# 2.    Plot the total_bill column histogram
df.hist(column='total_bill')
#the data here is divided into 10 bins, recorded in the x axis and its frequency is recorded in the y axis
```

```python
# 3.    Create a scatter plot presenting the relationship between total_bill and tip
bill_tip_scat=df.plot.scatter(x='total_bill',
                             y='tip')
bill_tip_scat
```

```python
# 4.    Create one image with the relationship of total_bill, tip and size.
import matplotlib.pyplot as mp
import seaborn as sb
data_plot=sb.heatmap(df.corr())
```

```python
# 5.    Present the relationship between days and total_bill value
days_total_bill=df[['total_bill','day']]
days_bill_df=days_total_bill.groupby(['day'])['total_bill'].agg(sum_total_bill='sum')
days_bill_df.plot()
```

```python
# 6.    Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

days_tip_scat=sb.scatterplot(x='tip', y='total_bill', hue='sex', data=df)
```

```python
#7.    Create a box plot presenting the total_bill per day differentiation the time (Dinner or Lunch)
sb.boxplot(x='day',y='total_bill',hue='time',data=df)
```

```python
# 8.    Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

lunch_tips=df[df['time']=='Lunch']['tip']
dinner_tips=df[df['time']=='Dinner']['tip']
fig, axes = mp.subplots(1, 2, figsize=(10,5))
sb.histplot(lunch_tips,ax=axes[0])
axes[0].set_title('Lunch Tips')
axes[0].set_xlabel('Tip')
axes[0].set_ylabel('Frequency')


sb.histplot(dinner_tips,ax=axes[1])
axes[1].set_title('Dinner Tips')
axes[1].set_xlabel('Tip')
axes[1].set_ylabel('Frequency')
```

```python
# 9.    Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker

sb.set_style("whitegrid")

# Create a scatterplot for Male
sb.scatterplot(x="total_bill", y="tip", hue="smoker", data=df[df["sex"] == "Male"])
mp.title("Male")
mp.show()

# Create a scatterplot for Female
sb.scatterplot(x="total_bill", y="tip", hue="smoker", data=df[df["sex"] == "Female"])
mp.title("Female")
mp.show()
```
