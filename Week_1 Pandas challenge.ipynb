```python
#url to the dataset:
#--> https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv



# Challenge Questions
# 1. Which was the most-ordered item?
# 2. For the most-ordered item, how many items were ordered?
# 3. What was the most ordered item in the choice_description column?
# 4. How many items were ordered in total?
# 5. Turn the item price into a float
# 6. How much was the revenue for the period in the dataset?
# 7. How many orders were made in the period?
# 8. What is the average revenue amount per order?
# 9. How many different items are sold?

import pandas as pd

url='https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
df=pd.read_csv(url,sep='\t')
df.head()   #check how the df looks
df.columns  #check the num of columns
df.info()   #check the datatypes 
```

```python
df
```

```python
#we can see that there are null values in choice description column. lets check this.
df['choice_description'].unique

```

```python
# 1. Which was the most-ordered item?
# 2. For the most-ordered item, how many items were ordered?
# 3. What was the most ordered item in the choice_description column?
# 4. How many items were ordered in total?
# 5. Turn the item price into a float
# 6. How much was the revenue for the period in the dataset?
# 7. How many orders were made in the period?
# 8. What is the average revenue amount per order?
# # 9. How many different items are sold?
# df_most_ordered=df.groupby(by=['item_name'])['order_id'].count()
df_most_ordered=df.groupby('item_name').agg(num_times_ordered=('order_id','nunique')).sort_values(['num_times_ordered'],ascending=False)
#aggregated the item name column and took the count of order_id for the column
```

```python
df_most_ordered
```

```python
# We can see that item names mentioned below are the same but differ by '-'.Lets name them in a uniform way.
# Chips and Fresh Tomato Salsa
# Chips and Tomatillo-Green Chili Salsa
# Chips and Tomatillo Green Chili Salsa
# Chips and Tomatillo Red Chili Salsa
# Chips and Tomatillo-Red Chili Salsa
# Chips and Roasted Chili-Corn Salsa
# Chips and Roasted Chili Corn Salsa
df['item_name'].replace({'Chips and Tomatillo-Green Chili Salsa':'Chips and Tomatillo Green Chili Salsa',
                                 'Chips and Tomatillo-Red Chili Salsa':'Chips and Tomatillo Red Chili Salsa',
                                 'Chips and Roasted Chili-Corn Salsa':'Chips and Roasted Chili Corn Salsa',
                                 'Chips and Fresh Tomato Salsa':'Chips and Fresh Tomato Salsa'},inplace=True)
df_most_ordered=df.groupby('item_name').agg(num_times_ordered=('order_id','nunique')).sort_values(['num_times_ordered'],ascending=False)
```

```python
df_most_ordered

# Chips and Tomatillo Green Chili Salsa	43
# Chips and Tomatillo-Green Chili Salsa	31
# Earlier, there were two entries for the chips and tomatillo salsa. Now we have a single row for the entry.
# So the texts have been replaced successfully.
```

```python
# 2. For the most-ordered item, how many items were ordered?
#We know that the most ordered item is Chicken Bowl.
filt=df['item_name']=='Chicken Bowl' #created a filter to sepcify name of the item
num_quantity_ordered=df[filt]['quantity'].sum() #calculted the sum of quantity related to the item chicken bowl
num_quantity_ordered
```

```python
#another way we could do this
new_df=df.groupby('item_name').filter(lambda x:'Chicken Bowl' in x['item_name'].values)
df2=new_df.groupby('item_name')['quantity'].sum()
df2
```

```python
# 3. What was the most ordered item in the choice description column
most_ordered_cd=df.groupby('choice_description').agg(num_times=('order_id','nunique')).sort_values(['num_times'],ascending=False)
most_ordered_cd

# --[Diet Coke]	
```

```python
# 4. How many items were ordered in total?

total_quantity_ordered=df['quantity'].sum()
total_quantity_ordered
```

```python
#5. Turn the item price into a float


#we can see that item_price has '$' symbol because of which we can't convert it into float directly
#First lets extract only the number part of the string
df['item_price']=df['item_price'].str.extract('(\d+\.\d+)').astype('float')
```

```python

df['item_price']='$'+df['item_price'].astype(str)
```

```python
# 6. How much was the revenue for the period in the dataset?

df['revenue']=df['item_price'].str.extract('(\d+\.\d+)',expand=False).astype(float)*df['quantity']
df['revenue'].sum()
```

```python
# 7. How many orders were made in the period?

df['order_id'].nunique()
```

```python
# 8. What is the average revenue amount per order?
df.groupby('order_id').agg('mean')['revenue']
```

```python
# 9. How many different items are sold?
df['item_name'].nunique()
```
